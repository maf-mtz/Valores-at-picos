## **AÑO 2020**

```python
import pandas as pd
data20= pd.read_excel('gastos_costos2020.xlsx') 
data20
```

```python
data20 = data20.drop(['Unnamed: 0','Columna1'], axis=1)
data20
```

```python
data20.info()
```

```python
valores_nulos=data20.isnull().sum()
valores_nulos
```

### Eliminar valores nulos

```python
#Reemplazamos valores nulos del dataframe
data1=data20.fillna(round(data20.mean()))
data1
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos

#solo no se eliminaron los valores nulos de las variables que son cualitativas
```

```python
cuantitativas_2020=data1.iloc[:,8:15]
cuantitativas_2020
```

### 1er Data Frame con desviación estándar

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize =(15, 8))
cuantitativas_2020.plot(kind='box', vert=False)
plt.title("Valores Atípicos de 2020")
plt.show() #dibujamos el diagrama
     
```

```python
y=cuantitativas_2020
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= cuantitativas_2020[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
df_20= cuantitativas_2020[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df_20
```

```python
valores_nulos=df_20.isnull().sum()
valores_nulos
```

```python
data_clean_2020=df_20.copy()
data_clean_2020=data_clean_2020.fillna(round(df_20.mean(),1))
data_clean_2020
```

```python
valores_nulos=data_clean_2020.isnull().sum()
valores_nulos
```

```python
#Descarga df limpio del año 2020 
data_clean_2020.to_csv("desv_2020.csv")
from google.colab import files 
files.download("desv_2020.csv")
```

### 1er Data Frame con cuantiles

```python
y=cuantitativas_2020

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= cuantitativas_2020[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data_clean_iqr_20= cuantitativas_2020[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr_20
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean_iqr_20.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
data_clean_iqr_2020=data_clean_iqr_20.copy()
data_clean_iqr_2020=data_clean_iqr_2020.fillna(round(data_clean_iqr_20.mean(),1))
data_clean_iqr_2020
```

```python
valores_nulos=data_clean_iqr_2020.isnull().sum()
valores_nulos
```

```python
#Descarga df limpio del año 2020 
data_clean_iqr_2020.to_csv("iqr_2020.csv")
from google.colab import files 
files.download("iqr_2020.csv")
```
