## **Aplicar métodos (Desviación Estándar y Rango Intercuartílico) para eliminar Outliers, comprobar con diagramas de caja por cada columna**

## Columna 2. ventas_precios_corrientes

```python
#diagrama de caja o bigote
#fig = plt.figure(figsize =(5, 3))
plt.boxplot(datos['ventas_precios_corrientes']) 
plt.title("Outliers de 'ventas_precios_corrientes'")
plt.show() 
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=datos["ventas_precios_corrientes"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
     
```

```python
#Encontramos Outliers
outliers_iqr= datos[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
data_clean_iqr_c2= datos[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr_c2
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean_iqr_c2["ventas_precios_corrientes"]) 
plt.title("Outliers de 'ventas_precios_corrientes'")
plt.show() #dibujamos el diagrama
```
